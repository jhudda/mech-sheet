Consider the piece of string the extends a distance $\delta x$ along the $x$ axis, and $\delta y$ along the $y$ axis. We can write:
$$ \delta y = f'(x) \delta x + O(\delta x^2)$$ 
We now consider the limit $\delta x \rightarrow 0$, such that the section is approximately straight. We can neglect the terms $O(\delta x^2)$ in the above, and also write the length of the infinitesimal piece as:
$$ds = \sqrt{dx^2 + f'(x)^2 dx^2} = dx \sqrt{1 + f'(x)^2}$$
We can then sum this up for all intervals $dx$ over the whole gap. Since said intervals are taken to be in the infinitesimal limit, the sum becomes an integral:
$$L = \int ds = \int_{-l/2}^{l/2} dx \sqrt{1 + f'(x)^2}$$

If we take the reference point of gravitational potential to be the line $y = 0$, then the potential energy of the entire string may be calculated by summing up, for each infinitesimal segment of string, the mass of the segment multiplied by the displacement of it from the line $y = 0$ multiplied by gravitational acceleration $g$. The first of these terms may be written as $\rho ds$, where $ds$ is as defined earlier. The second of these is given simply by $f(x)$. The final term is simply $g$. Then, the infinitesimal contribution is:
$$dV = \rho g f(x) ds = \rho g f(x) \sqrt{1 + f'(x)^2} dx$$
Summing these up over the whole string:
$$V = \int dV = \rho g \int_{-l/2}^{l/2} dx \, f(x) \sqrt{1 + f'(x)^2} 

The string will reach stable equilibrium at the position with potential energy, so to find the shape of the string, we need to find the $f(x)$ such that the above $V$ is minimised. We may notice however, that we could simply take this to meaning the string droops an infinite distance down from the edges, since this would obviously lower the potential energy an infinite amount. It becomes apparent that we must impose a constraint on the equation $f(x)$, that being that the length relation detailed above holds. 

This problem can be neatly encapsulated in minimising the functional $S[f(x)]$:
$$S[f(x)] = V - \rho g \lambda \left(\int_{-l/2}^{l/2} dx \sqrt{1 + f'(x)^2} - L\right)$$
To minimise this, in addition to no first order change under small adjustments to $f(x)$, we also require $\partial S / \partial \lambda = 0$, which ensures that the bracketed term on the right is 0, such that the constraint on the length is definitely satisfied. Here, $\lambda$ is referred to as a Lagrange multiplier.

Now, consider a more general functional, again called $S[f(x)]$:
$$S[f(x)] = \int_{-l/2}^{l/2} dx \,\mathcal{L} (f(x), f'(x)) $$
The Euler-Lagrange equations tell us that the $f(x)$ that extremises this must satisfy:
$$\frac{\partial \mathcal{L}}{\partial f} -  \frac{d}{dx}\left( \frac{\partial \mathcal{L}}{\partial f'}\right)=0$$
Now multiply all terms by $f'$, and add on and subtract $f'' (\partial \mathcal{L} / \partial f')$:
$$f'\frac{\partial \mathcal{L}}{\partial f} + f'' \frac{\partial \mathcal{L}}{\partial f'} - f'' \frac{\partial \mathcal{L}}{\partial f'} - f' \frac{d}{dx}\left( \frac{\partial \mathcal{L}}{\partial f'}\right)=0$$
$$ \left(\frac{\partial \mathcal{L}}{\partial f} \frac{\partial f}{\partial x} +  \frac{\partial \mathcal{L}}{\partial f'} \frac{\partial f'}{\partial x}\right) - \frac{d}{dx}\left(f' \frac{\partial \mathcal{L}}{\partial f'} \right) = 0$$
$$\frac{d \mathcal{L}}{dx} - \frac{d}{dx}\left(f' \frac{\partial \mathcal{L}}{\partial f'} \right) = \frac{d}{dx} \left(\mathcal{L} - f' \frac{\partial \mathcal{L}}{\partial f'}\right) = 0$$

Which implies that the last quantity in the brackets is independent of $x$, so may be set as equal to some constant $b$.
